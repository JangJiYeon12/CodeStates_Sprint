- mean : 232.2

```python
import pandas as pd
import numpy as np

def mymean(v):
    return sum(v) / len(v)

def myvar(v):
    return sum((v - mymean(v))**2) / len(v)

def mystd(v):
    return np.sqrt(myvar(v))

v = np.array([243, 278, 184, 249, 207])
```
